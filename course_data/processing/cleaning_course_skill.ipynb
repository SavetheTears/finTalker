```python
for i in range(len(df)):
    ans_soft = df.Course_Soft_Skills[i]
    if ans_soft == None:
        continue
    stemmer = ans_soft.split()
    if stemmer[0] == 'No' or stemmer[0] == 'no.':
        df.Course_Soft_Skills[i] = None
        continue
        
for i in range(len(df)):
    ans_hard = df.Course_Hard_Skills[i]
    if ans_hard == None:
        continue
    stemmer = ans_hard.split()
    if stemmer[0] == 'No' or stemmer[0] == 'no.':
        df.Course_Hard_Skills[i] = None
        continue
```

```python
for i in range(len(df)):
    ans_hard = df.Course_Hard_Skills[i]
    if ans_hard == None:
        continue
        
    if 'Hard skills not mentioned' in ans_hard or "Hard skills mentioned: No." in ans_hard or \
    'Hard skills mentioned: None' in ans_hard or 'No hard skills mentioned' in ans_hard or \
    'Hard skills mentioned: no' in ans_hard:
        df.Course_Hard_Skills[i] = None
        
    if 'but none are specified' in ans_hard:
        df.Course_Hard_Skills[i] = NOT_SPECIFIED
```

```python
# delete over verbose word. Hard
for i in range(len(df)):
    ans_hard = df.Course_Hard_Skills[i]
    if ans_hard == None or ans_hard ==NOT_SPECIFIED:
        continue
    
    # contain :
    type_1 = df.Course_Hard_Skills[i].split(': ',1)
    if len(type_1) == 2:
        df.Course_Hard_Skills[i] = type_1[1]
        continue
        
    # 'include'
    type_2 = df.Course_Hard_Skills[i].split('include ',1)
    if len(type_2) == 2:
        df.Course_Hard_Skills[i] = type_2[1]
        continue
    
    # -
    type_3  = df.Course_Hard_Skills[i].split('- ',1)
    if len(type_3) == 2:
        df.Course_Hard_Skills[i] = type_3[1]
        continue
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df.Course_Hard_Skills)
```

```python
# delete over verbose word. Soft
for i in range(len(df)):
    ans_soft = df.Course_Soft_Skills[i]
    if ans_soft == None or ans_soft ==NOT_SPECIFIED:
        continue
    
    # contain :
    type_1 = df.Course_Soft_Skills[i].split(': ',1)
    if len(type_1) == 2:
        df.Course_Soft_Skills[i] = type_1[1]
        continue
        
    # 'include'
    type_2 = df.Course_Soft_Skills[i].split('include ',1)
    if len(type_2) == 2:
        df.Course_Soft_Skills[i] = type_2[1]
        continue
    
    # -
    type_3  = df.Course_Soft_Skills[i].split('- ',1)
    if len(type_3) == 2:
        df.Course_Soft_Skills[i] = type_3[1]
        continue
        
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df.Course_Soft_Skills)
```

```python
import pandas as pd
NOT_SPECIFIED = 'Not_Specified'

df = pd.read_csv('./output_cleaned.csv')

for i in range(len(df)):
    if pd.isna(df.Course_Soft_Skills[i]):
        df.Course_Soft_Skills[i] = None
    
    if pd.isna(df.Course_Hard_Skills[i]):
        df.Course_Hard_Skills[i] = None
```

```python
for i in range(len(df)):
    ans_soft = df.Course_Soft_Skills[i]
    if not ans_soft or ans_soft ==NOT_SPECIFIED:
        continue
        
    if ans_soft == 'no' or 'no reference in the sentence' in ans_soft:
        df.Course_Soft_Skills[i] = None
        continue
        
    if '(Soft skills mentioned)' in ans_soft:
        df.Course_Soft_Skills[i] = df.Course_Soft_Skills[i][:-23]
```
